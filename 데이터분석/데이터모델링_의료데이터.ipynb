```python
# 변수선택(서비스의 대상을 기준으로 선택)
df2 = df[['성별','신장','체중','흡연여부','연령','혈액형','직업','재발여부']]
```

```python
# 결측치 제거/ 
df3 = df2.dropna()
```

```python
df3.shape
```

## 설명변수(독립변수) 목표변수(종속변수) 설정

```python
# 더미변수 처리 (문자데이터 → One Hot Encoding)
X = df3.drop(columns="재발여부")
X1 = pd.get_dummies(X) #더미화 시키기 
X1
```

```python
Y = df3["재발여부"]
Y
```

```python
# Decision Tree 
# 훈련용 데이터 
from sklearn.model_selection import train_test_split 

# 의사결정 나무 
from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# 검증 
from sklearn.metrics import classification_report

# 시각화 
from sklearn.tree import plot_tree
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X1,Y,test_size=0.3,
                                                random_state=1234)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
pipe_list = [('scaler',MinMaxScaler()),('model',DecisionTreeClassifier())]
pipe_model = Pipeline(pipe_list)
pipe_model
```

```python
hyper_list = {'model__max_depth':range(2,10),
             'model__min_samples_leaf':range(2,10),
             'model__criterion':['gini','entorpy'],
             'model__class_weight':[None,'balanced'],
              'model__min_samples_split':range(2,10),
             }

grid_model = GridSearchCV(pipe_model,param_grid=hyper_list,scoring='f1',
                         n_jobs= -1,
                         cv = 5)
grid_model.fit(X_train,Y_train)
```

```python
best_model = grid_model.best_estimator_
```

```python
Y_train_pred = best_model.predict(X_train)
Y_test_pred = best_model.predict(X_test)
```

## 분류 모델에서 확인하고자 하는 Target 항목의 데이터 비율이 적으면 
    * recall 값이 매우 떨어지는 현상이 도출된다. 

```python
print(classification_report(Y_train,Y_train_pred))
```

```python
# 분류트리 그려보기 
print(classification_report(Y_test,Y_test_pred))
```
