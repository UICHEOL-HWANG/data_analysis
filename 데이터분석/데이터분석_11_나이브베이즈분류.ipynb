## 나이브베이즈 분류 
* 베이즈 이론(베이지안 확률) 
    * A가 일어 났을 때 B가 일어날 확률 
    * 메일 제목에 '무료'라는 단어가 들어가 있을 때 , 스팸 메일일 확률 
    * 주관적으로 확률을 해석 
    * 데이터를 통해서 확률을 추정할 때 현재 관측된 데이터의 빈도만으로 분석하는 것이 아니라 분석자의 사전지식(혹은 알려진 사실, 분석자의 주관)까지 포함하여 분석하는 방법 
    * 베이즈 정리에서 확률이란, '주장' 혹은 '믿음의 신뢰도'로 나타난다.
* 나이브 베이즈 분류 모델은 베이즈 정리를 기반으로 한 지도학습 모델로, 스팸 메일 필터링, 텍스트 분류 등에 사용
* 나이브 베이즈는 데이터의 모든 특성 변수가 서로 동등하고 독립적이라는 가정 하에 분류를 실행한다. 
* 독립변수의 특성들은 나이브 베이즈 분류기에서 서로 연관성이 없고, 각각의 특성이 독립적으로 기여하는 것으로 간주한다.

```python
def ages(x):
    if x < 20 :
        return '10s'
    elif 20 <= x <30 :
        return '20s'
    elif 30 <= x < 40 :
        return '30s'
    elif 40 <= x < 50:
        return '40s'
    elif 50 <= x < 60:
        return '50s'
    elif 60 <= x < 70 :
        return '60s'
```

```python
data2.loc[:,"Age"] = data2["Age"].apply(ages)
```

```python
X = data.drop(["Ad Topic Line","City","Country","Timestamp","Clicked on Ad"],axis=1) # 독립변수 
y = data["Clicked on Ad"] # 종속변수 
```

```python
X
```

```python
X = pd.get_dummies(X,drop_first=True)
X
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
ss = StandardScaler() 
X_scaled = ss.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled,columns=X.columns)
X_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=7)
```

## 나이브 베이즈 예측 및 평가 

```python
from sklearn.naive_bayes import GaussianNB , BernoulliNB , CategoricalNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
nbc =GaussianNB()
nbc.fit(X_train,y_train)
nbc_pred = nbc.predict(X_test)
print(accuracy_score(y_test,nbc_pred))
print(confusion_matrix(y_test,nbc_pred))
print(classification_report(y_test,nbc_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth=3,random_state=7)
dtc.fit(X_train,y_train)
dtc_pred = dtc.predict(X_test)

print(accuracy_score(y_test,dtc_pred))
print(confusion_matrix(y_test,dtc_pred))
print(classification_report(y_test,dtc_pred))
```

```python
rtc = RandomForestClassifier(n_estimators=1000,random_state=7,max_depth=3)
rtc.fit(X_train,y_train)
rtc_pred = rtc.predict(X_test)

print(accuracy_score(y_test,rtc_pred))
print(confusion_matrix(y_test,rtc_pred))
print(classification_report(y_test,rtc_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
glm = LogisticRegression()
glm.fit(X_train,y_train)
glm_pred = glm.predict(X_test)

print(accuracy_score(y_test,glm_pred))
print(confusion_matrix(y_test,glm_pred))
print(classification_report(y_test,glm_pred))
```
