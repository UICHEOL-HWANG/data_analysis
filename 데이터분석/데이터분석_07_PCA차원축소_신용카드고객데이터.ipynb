```python
# 아웃라이어로 이상값인지 아닌지 판별하기
def outlier(x):
    x = x.describe()
    out_max = x.loc["75%"] + (1.5*(x.loc["75%"]-x.loc["25%"]))
    out_min = x.loc["25%"] - (1.5*(x.loc["75%"]-x.loc["25%"]))
    ol_result = pd.DataFrame([out_max,out_min],columns=out_max.index,index=["상한값","하한값"])
    result = pd.concat([x,ol_result])
    return result
```

```python
outlier(data)
# max값이 상한가에 잘 있으면 이상값이 잘 없는거다 
```

```python
def c_ratio(x):
    for idx,item in zip(x.index,x):
        if idx[1] == 0:
            unpaied = item 
        else :
            paid = item
            class_ratio = paid / (paid+unpaied) * 100 
            print(idx,f'{class_ratio:.2f}%')
```

## 분석을 위한 독립 변수 종속 변수 나누기 

```python
X = data.drop(["ID","정상지불예측"],axis=1)
y = data["정상지불예측"]
```

# 정규화 
* 독립변수들 사이에 변수간 단위차가 크기 때문에, 표준화,정규화를 해야 할 필요가 있다. (스케일링)
* 스케일러 
1) StandardScaler(기본 스케일러,평균과 표준편차를 사용): 평균을 제거 하고 데이터를 단위분산으로 조정, 그러나 이상치가 있다면 평균과 표준 편차에 영향을 미쳐 변환된 데이터의 확산은 매우 달라지게 된다. 따라서 이상치가 있는 경우 균형 잡힌 척도를 보장할 수 없다.** 

2) MinMaxScaler(최대,최소값이 각각 1,0이 되도록 스케일링) : 모든 Feature(독립변수) 값이 0~1사이에 있도록 데이터를 재 조정. 다만 이상치가 있는 경우 변환된 값이 매우 좁은 범위로 압축될 수 있다. 즉 MinMaxScaler 역시 이상치에 매우 민감하다. **

3) MaxAbsScaler(최대 절대값과 0이 각각 1,0이 되도록 스케일링 작업) : 절대값이 0-1사이에 매핑되도록함. -1~1 사이로 재조정한다. 양수 데이터로만 구성된 특징 데이터 셋에서는 MinMaxScaler와 매우 유사하게 작동하며, 큰 이상치에 민감할 수 있음. **

4) RobustScaler(중앙 값(median)과 IQR(4분위수)를 사용 하여 outlier의 영향을 최소화 하는 스케일러): outlier의 영향을 축소화한 기법이다. 중앙값(median)과 IQR(4분위수)를 사용하기 때문에, StandardScaler와 비교해보면 표준화 후 동일한 값을 더 넓게 분포시키고 있음을 확인할 수 있다. **



```python
from sklearn.preprocessing import RobustScaler # RobustScaler 소환
```

```python
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
pd.DataFrame(X_scaled,columns=X.columns) # 정규화 완성 
```

```python
X_scaled = pd.DataFrame(X_scaled,columns=X.columns) 
```

```python
X_scaled.head()
```

## 홀드아웃 


```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.3,random_state=10) #데이터 분리 완료 
```

## 알고리즘 불러와서 분석하기 

```python
# 로지스틱 회귀분석 
from sklearn.linear_model import LogisticRegression #로지스틱 회귀분석 
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix #결과 
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
pred = lr.predict(X_test)
print(confusion_matrix(y_test,pred))
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
```

```python
y.value_counts() #불균형이 심해 정밀도 값과 F1스코어 값이 작음 
```

## PCA를 통해 변수 축소
* 축소 후 로지스틱 회귀분석 실시 
* n_components=0.80 :  전체의 80% 설명가능한 변수까지 선택 
* n_components=none : 모든 주성분의 분산비율 확인 가능 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.80)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca.n_components_ #전체 23개가 나옴 
```

```python
X_pca[0]
```

## PCA를 통해서 6개로 차원축소한 데이터로 분석 

```python
X_train2,X_test2,y_train2,y_test2 = train_test_split(X_pca,y,test_size=0.3,random_state=10) # 차원축소 후 데이터 분리 완료 
```

```python
lr2 = LogisticRegression()
lr2.fit(X_train2,y_train2)
pred2 = lr2.predict(X_test2)
print(confusion_matrix(y_test2,pred2))
print(accuracy_score(y_test2,pred2))
print(classification_report(y_test2,pred2))
```
