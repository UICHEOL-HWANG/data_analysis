```python
def add_genre_name(j):
    genres = []
    ar = json.loads(j)
    for a in ar:
        genres.append(a.get("name"))
    return " ".join(genres)
        
```

```python
def lang_get(x):
    lang = [] 
    lan = json.loads(x)
    for i in lan:
        lang.append(i.get("iso_639_1"))
    return ",".join(lang)
```

```python
def key_get(x):
    keyword = []
    key = json.loads(x)
    for i in key:
        keyword.append(i.get("name"))
    return " ".join(keyword)
```

```python
data["genres"] = data["genres"].apply(lambda x: add_genre_name(x))
```

```python
data["spoken_languages"] = data["spoken_languages"].apply(lambda x:lang_get(x))
```

```python
data["keywords"] = data["keywords"].apply(lambda x: key_get(x))
```

```python
data["production_companies"].apply(lambda x: key_get(x))
```

```python
df = data[["id","title","genres","vote_average","popularity","keywords","overview"]]

df
```

## genres의 단어 벡터화  / 코사인유사도 이용 
* 비슷한 장르를 가진 행을 찾고 , 유사도가 높은 행을 기준으로 내림차순 정렬


```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range=(1,2))
genre_mat = cv.fit_transform(df["genres"])
print(genre_mat)
```

```python
genre_mat.shape
```

## 코사인 유사도 사용 
* genre_mat의 유사도 산출 후 , genre_sim에 저장 
* 비슷한 장르 영화를 찾기 위해 계산 
* 코사인 유사도는 주로 문자, 문서의 유사도 계산에 사용됨

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
genre_sim = cosine_similarity(genre_mat,genre_mat)
print(genre_sim.shape)
print(genre_sim[:2])
```

```python
def find_sim_movie(df,sorted_genre_sim,title_name,top_n=10):
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    similar_indexes = sorted_genre_sim[title_index, :(top_n)]
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)
    return df.iloc[similar_indexes]
```

```python
find_sim_movie(df,sorted_gerne_sim,"Superman III",10)
```
