````python
np.random.seed(19680801)
data = {
    "a" : np.arange(50),
    "c" : np.random.randint(0,50,50),
    "d" : np.random.randn(50)
}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig,ax = plt.subplots(figsize=(5,2.7),layout='constrained')
ax.scatter('a','b',c='c',s='d',data=data) # 산점도 그래프 만들기
# c = 찍히는 점의 색 , s = 점의 크기

ax.set_xlabel("A값")
ax.set_ylabel("B값",rotation=0,labelpad=15)
````

## Matplotlib 2가지 코딩 스타일

## 1) 00-스타일 : matplotlib 공식 추천 스타일 
* 복잡한 그래프를 그릴 때 사용 
* 사용법이 pyplot스타일보다 복잡함 


````python
x = np.linspace(0,2,100)

fix,ax = plt.subplots(figsize=(5,2.7),
                     layout = "constrained"
                     )

ax.plot(x,x,label="linear")
ax.plot(x,x**2,label="quadratic")
ax.plot(x,x**3,label="cubic")
ax.set_xlabel("x축")
ax.set_ylabel("y축",rotation=0)
ax.set_title("샘플 라인차트")
ax.legend()
````

## 2) pyplot 스타일 
* 00-style보다 사용법이 더 간단 
* 복잡한 그래프를 그릴 때는 적합하지 않음

````python
x = np.linspace(0,2,100)

plt.figure(figsize=(5,2.7),
                     layout = "constrained"
                     )

plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quadratic")
plt.plot(x,x**3,label="cubic")
plt.xlabel("x축")
plt.ylabel("y축",rotation=0)
plt.title("샘플 라인차트")
plt.legend()
````

## 그래프 꾸미기 

### 1) 선 모양 바꾸기 set_linestyle(모양)

종류 
* linestyle or ls: {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}

````python
# 4개의 랜덤 데이터셋 만들기 
data1,data2,data3,data4 = np.random.randn(4,100)
````

````python
fig,ax = plt.subplots(figsize=(5,2.7))
x = np.arange(len(data1))
ax.plot(x,np.cumsum(data1),color='blue',linewidth=3,linestyle="--")
l, = ax.plot(x,np.cumsum(data2),color='orange',linewidth=2)
l.set_linestyle(':')
plt.show()
````

## 2) 색바꾸기 (facecolor,edgecolor)

````python
fig,ax = plt.subplots(figsize=(5,2.7))
ax.scatter(data1,data2,s=50,facecolor='C8',edgecolor='k')
plt.show()

# C컬러 색에 따라 0~10까지 있는 듯
````

## 선 넓이, 선 형태 마커 모양과 크기 



````python
fig,ax = plt.subplots(figsize=(5,2.7))

ax.plot(data1,'o',label='data1') # o 동그라미 
ax.plot(data2,'d',label='data2') # d 다이아 
ax.plot(data3,'v',label='data3') # v 브이 
ax.plot(data4,'s',label='data4') # s 네모 
ax.legend()
plt.show()
````

## 4) x축 , y축 이름,눈금,그래프 이름 

````python
mu,sigma = 115,15 
x = mu + sigma * np.random.randn(10000)
fig,ax = plt.subplots(figsize=(5,2.7),layout="constrained")
n, bins, patches = ax.hist(x,50,density=True,facecolor ="C0",alpha=0.75)


ax.set_xlabel("길이[cm]")
ax.set_ylabel("확률")
ax.set_title("땅돼지의 길이 (실제는 아님)")
ax.text(75,.025,r'$\mu=115,\ \sigma=15$')
ax.axis([55,175,0,0.03])
ax.grid(True)
````
