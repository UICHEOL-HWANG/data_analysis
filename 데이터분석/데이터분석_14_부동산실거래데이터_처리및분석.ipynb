```python
data = pd.read_csv(r"C:\Users\hi\Desktop\bigdata\전국부동산실거래가\전국부동산실거래가\2006_01.csv",encoding="cp949",sep=",",skiprows=15)
```

```python
dataframes = []

for file_path in file_list:
    try:
        # 빈 행이나 이상한 문자를 스킵하고 데이터 불러오기
        data = pd.read_csv(file_path, encoding="cp949", sep=",", skiprows=15)
        
        # 데이터가 비어있는 경우 건너뛰기
        if not data.empty:
            dataframes.append(data)
        else:
            print(f"데이터가 비어있는 파일: {file_path}")
    except Exception as e:
        print(f"파일 불러오기 오류: {file_path}, 에러 메시지: {str(e)}")

# 리스트에 있는 모든 DataFrame을 concat 함수로 병합
merged_data = pd.concat(dataframes, ignore_index=True)
```
