## 데이터프레임에서 조건에 맞는 행만 갖고오기
* 데이터프레임[데이터프레임["컬럼명"] 조건식] → 조건식이 참인 것만 가져옴
* \>,< , >=, <=,==,!=,&(and), |(or)

````python
# 조건이 1개인 경우
df3[df3["범죄율"] <= 0.02729]
````

````python
# 0.05 < 범죄율 < 0.01 2개의 조건

df3[(df3["범죄율"] > 0.05) | (df3["범죄율"] < 0.1)]
````

````python
# 조건이 4개인경우 
# 범죄율이 0.05미만, 가구당 평균 방 수 >=4,하위 계층비율 3 이하인 집
````

````python
df3[(df3["범죄율"] < 0.05) & (df3["가구당평균방수"] >= 4) & (df3["하위계층비율"] <= 3) ]
````
