```python
import requests
import pandas as pd 
import numpy as np 
from bs4 import BeautifulSoup as bs 
import time
from tqdm import tqdm
from IPython.display import clear_output
```

## 보배드림 중고차 파싱 
* 첫 번째 기본 사이트 794페이지 크롤링

```python
data_list2 = []

for pages in tqdm(range(745,747)):

    url = f"https://www.bobaedream.co.kr/cyber/CyberCar.php?sel_m_gubun=ALL&page={pages}&order=S11&view_size=70"
    print(f"현재 {pages} 페이지 크롤링 중..")
    r = requests.get(url)
    soup = bs(r.text,"lxml")
    time.sleep(5)

    for i in soup.select(".product-item"):
        data_dict2 ={}
        data_dict2["title"] = i.select_one(".tit.ellipsis").text.strip()
        data_dict2["link"] = i.select_one(".mode-cell.thumb>a")["href"]
        data_dict2["thumb"] = i.select_one(".mode-cell.thumb>a>img")["src"]
        data_dict2["desc"] = i.select_one(".stxt.ellipsis").text
        data_dict2["option"] = [a.text for a in i.select(".data.clearfix>.data-item")]
        data_dict2["year"] = i.select_one(".mode-cell.year").text.strip()
        data_dict2["fuel"] = i.select_one(".mode-cell.fuel>.text").text
        data_dict2["km"] = i.select_one(".mode-cell.km>.text").text
        data_dict2["price"] = i.select_one(".price-whole.cr").text if i.select_one(".price-whole.cr") is not None else None
        data_dict2["sell-info"] = i.select_one(".list-seller>.seller-name").text.strip() if  i.select_one(".list-seller>.seller-name") is not None else None
        data_dict2["content_info"] = [a.text.replace("\n"," ") for a in i.select(".content-list")]
        data_list2.append(data_dict2)
    clear_output(wait=True)
```

## 크롤링 결과 

* 누락값이 너무 많았음 (팔린 차들은 가격 블라인드처리)
* 누락값 4만여개 이상 / 총 데이터 수 5만개 
* 클래스 불균형 예상되어 다른 값으로 대체

```python
datas = ["330","111","213","118","2","3","2","5","44","53","7"]
data_list_sec = []
for links in datas:
    for pages in tqdm(range(1,40)):
        try:
            print(f"현재 {links}영역 중 {pages}페이지 크롤링 중")
            base_url = f"https://www.oasis.co.kr/product/list?categoryId={links}&page={pages}&sort=priority&direction=desc&couponType=&rows=60"
            header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"}
            r2 = requests.get(base_url,headers=header,verify=False)
            time.sleep(3)
            soup2 = bs(r2.text,"lxml")
            time.sleep(5)
            for i in soup2.select(".wrapBox"):
                data_dict_sec = {}
                data_dict_sec["link"] = i.select_one(".wrapImg>a")["href"] if i.select_one(".wrapImg>a") is not None else None
                data_dict_sec["img"] = i.select_one(".wrapImg>a>img")["src"] if i.select_one(".wrapImg>a>img") is not None else None
                data_dict_sec["title"] = i.select_one(".info_title>.innerBox>.inner").text.strip().replace("\n","").replace("\t","").replace(" ","") if  i.select_one(".info_title>.innerBox>.inner") is not None else None
                data_dict_sec["delivery"] = i.select_one(".info_badges").text.strip() if  i.select_one(".info_badges") is not None else None
                data_dict_sec["price"] = i.select_one(".price_discount").text if i.select_one(".price_discount") is not None else 0
                data_dict_sec["percent"] = i.select_one(".info_price>span").text if i.select_one(".info_price>span") is not None else 0 
                data_dict_sec["original_price"] = i.select_one(".info_price .price_original").text if i.select_one(".info_price .price_original") is not None else 0 
                data_list_sec.append(data_dict_sec)
        except Exception as e :
            print(f"페이지 없음 {e}")
            time.sleep(5)
            pass
        clear_output(wait=True)
```

```python
df_real = pd.DataFrame(data_list_sec)
```
