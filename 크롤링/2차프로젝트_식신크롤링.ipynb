```python
all_pages = []

for page in local:
    driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install())) #웹 드라이버 버전에 맞게 다운 받고 옵션을 추가해줌 
    driver.set_window_size(1920, 1080) # 웹 브라우저 해상도 조절 
    print(f"현재 {page} 크롤링 작업중..")
    driver.get(f"https://www.siksinhot.com/search?keywords={page}") # 크롤링할 웹사이트 주소  
    
    time.sleep(3)
    
    while True:
        try:
            next_button = driver.find_element(By.CSS_SELECTOR, "#main_search > div > article > section > div.div_search_menu > div:nth-child(1) > a")
            next_button.click()
            time.sleep(5)
        except Exception as e:
            print("버튼1 없다. 다음")
            try:
                next_button2 = driver.find_element(By.CSS_SELECTOR, "#main_search > div > article > section > div.div_search_menu > div:nth-child(2) > a")
                next_button2.click()
                time.sleep(5)  # 버튼2 클릭 후 대기
            except Exception as e:
                print("버튼2 없다. 다음")
                html_page = driver.page_source
                all_pages.append(html_page)
                break
    
    driver.quit()
```

```python
with open("식신크롤링.pkl", "wb") as file:
    pickle.dump(all_pages, file)
```
