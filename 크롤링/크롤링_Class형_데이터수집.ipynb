```python
import requests
import pandas as pd 
from bs4 import BeautifulSoup as bs
import time
```

```python
class NewsCrawling:
    def __init__(self,query,header):
        self.url ="https://search.naver.com/search.naver?where=news&query="
        self.query = query
        self.header = header
        
    def search_news(self,start_page):
        url = self.url+self.query+f"&start={start_page}"
        header = {"User_Agent":self.header}
        r = requests.get(url,headers=header)
        soup = bs(r.text,"lxml")
        for i in soup.select(".news_area"):
            data_dict = {}
            data_dict["desk"]=i.select_one(".info_group").text.strip().split()[0].split("언론사")[0]
            data_dict["news_time"]=i.select_one(".info_group>a+span").text
            data_dict["title"]=i.select_one(".news_tit").text
            data_dict["link"] = i.select_one(".news_tit")["href"]
            data_dict["desc"]=i.select_one(".dsc_wrap").text.strip()
        return data_dict
```

```python
all_data = []
input_search = ["이선균","전혜진"]
for keyword in input_search:
    for num in range(1,1000,10):
        time.sleep(2)
        news = NewsCrawling(keyword,"Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36")
        all_data.append(news.search_news(num))
        print(f"현재 {keyword} 내용의 {num} 페이지 수집중 ....",end="\r")
```
