```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
import time
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install())) #웹 드라이버 버전에 맞게 다운 받고 옵션을 추가해줌 
driver.set_window_size(1920,1080)# 웹 브라우저 해상도 조절 
driver.get("https://shopping.naver.com/home")
```

```python
search_box = driver.find_element(By.CSS_SELECTOR,"#gnb-gnb > div._gnb_header_area_150KE > div > div._gnbLogo_gnb_logo_3eIAf > div > div._gnbSearch_gnb_search_3O1L2 > form > div._gnbSearch_inner_2Zksb > div._searchInput_search_input_QXUFf > input")
search_box.send_keys("꿀고구마"+Keys.ENTER)
```

```python
while True:
    # 현재 스크롤 위치 저장
    last_height = driver.execute_script("return document.body.scrollHeight")

    # 자바스크립트를 사용하여 스크롤을 끝까지 내림
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 스크롤이 끝까지 내려갈 때까지 대기
    time.sleep(2)  # 적절한 대기 시간을 설정해야 함

    # 새로운 스크롤 위치
    new_height = driver.execute_script("return document.body.scrollHeight")

    # 스크롤이 끝까지 도달했는지 확인
    if new_height == last_height:
        break  # 스크롤이 끝까지 내려갔으면 루프 종료

# 페이지의 내용을 가져오기
page_html = driver.page_source
```

```python
soup = bs(page_html,"lxml")
```

```python
a = soup.select(".adProduct_info_area__dTSZf")
b = soup.select(".product_inner__gr8QR")

# 모든 데이터를 저장할 리스트 초기화
all_list = []

# 첫 번째 선택자에서 데이터 추출하여 리스트에 추가
for i in a:
    title = i.select_one(".adProduct_link__NYTV9.linkAnchor").text.strip()
    price = i.select_one(".price_num__S2p_v").text.strip()
    delivery_fee = i.select_one(".price_delivery__yw_We").text.strip()
    category = i.select_one(".adProduct_depth__s_IUT").text.strip()
    delivery_info = i.select_one(".deliveryInfo_tag__M4MGK").text.strip() if i.select_one(".deliveryInfo_tag__M4MGK") is not None else "None"
    bookmark = i.select_one(".adProduct_num__t7R1x").text
    all_list.append([title, price, delivery_fee, category, delivery_info,bookmark])

# 두 번째 선택자에서 데이터 추출하여 리스트에 추가
for k in b:
    title = k.select_one(".product_link__TrAac.linkAnchor").text.strip()
    price = k.select_one(".price_num__S2p_v").text.strip()
    delivery_fee = k.select_one(".price_delivery__yw_We").text.strip()
    category = k.select_one(".product_depth__I4SqY").text.strip()
    delivery_info = k.select_one(".deliveryInfo_tag__M4MGK").text.strip() if k.select_one(".deliveryInfo_tag__M4MGK") is not None else "None"
    bookmark = k.select_one(".product_num__fafe5").text
    all_list.append([title, price, delivery_fee, category, delivery_info,bookmark])

# 데이터프레임 생성
import pandas as pd
df = pd.DataFrame(all_list, columns=["제목", "가격", "배송비", "카테고리", "배송옵션","찜하기"])
```

```python
a = soup.select(".adProduct_info_area__dTSZf")
b = soup.select(".product_inner__gr8QR")

# 데이터를 저장할 리스트 초기화
titles, prices, delivery_fees, categories, delivery_infos = [], [], [], [], []

# 첫 번째 선택자에서 데이터 추출하여 리스트에 추가
for i in a:
    title = i.select_one(".adProduct_link__NYTV9.linkAnchor").text
    price = i.select_one(".price_num__S2p_v").text.strip()
    delivery_fee = i.select_one(".price_delivery__yw_We").text.strip()
    category = i.select_one(".adProduct_depth__s_IUT").text.strip()
    delivery_info = i.select_one(".deliveryInfo_tag__M4MGK").text.strip() if i.select_one(".deliveryInfo_tag__M4MGK") is not None else "None"
    
    titles.append(title)
    prices.append(price)
    delivery_fees.append(delivery_fee)
    categories.append(category)
    delivery_infos.append(delivery_info)

# 두 번째 선택자에서 데이터 추출하여 리스트에 추가
for k in b:
    title = k.select_one(".product_link__TrAac.linkAnchor").text
    price = k.select_one(".price_num__S2p_v").text.strip()
    delivery_fee = k.select_one(".price_delivery__yw_We").text.strip()
    category = k.select_one(".product_depth__I4SqY").text.strip()
    delivery_info = k.select_one(".deliveryInfo_tag__M4MGK").text.strip() if k.select_one(".deliveryInfo_tag__M4MGK") is not None else "None"
    
    titles.append(title)
    prices.append(price)
    delivery_fees.append(delivery_fee)
    categories.append(category)
    delivery_infos.append(delivery_info)

# 데이터를 딕셔너리에 추가
data = {
    "Title": titles,
    "Price": prices,
    "Delivery Fee": delivery_fees,
    "Category": categories,
    "Delivery Info": delivery_infos
}

# 데이터프레임 생성
import pandas as pd
df = pd.DataFrame(data)
```

```python
# 선택자로부터 데이터 추출
item = soup.select(".basicList_list_basis__uNBZx")

data_list = []

# 선택자로부터 데이터를 추출하여 딕셔너리에 추가하고 리스트에 넣기
for i in item:
    ad_product_items = [",".join(i.text.split(",")) for i in i.select(".adProduct_item__1zC9h .adProduct_link__NYTV9.linkAnchor")]
    product_items = [i.text for i in i.select(".product_item__MDtDF .product_link__TrAac.linkAnchor")]
    price = [i.text for i in i.select(".price_num__S2p_v")]
    delivery = [i.text for i in i.select(".price_delivery__yw_We")]
    fast_delivery = [i.text for i in i.select(".fastDeliveryInfo_delivery_arrival__Dy12_")]
    category1 = [i.text for i in i.select(".adProduct_depth__s_IUT")]
    category2 = [i.text for i in i.select('.product_depth__I4SqY')]
    agriculture1 = [i.text for i in i.select(".adProduct_detail_box___4Crv")]
    agriculture2 = [i.text for i in i.select(".product_detail_box__QjZP8")]

    # 딕셔너리에 데이터 추가
    data_list.append([ad_product_items+product_items,price,delivery,fast_delivery,category1+category2,agriculture1+agriculture2])
    # 딕셔너리를 리스트에 추가
    df = pd.DataFrame(data_list)
```
