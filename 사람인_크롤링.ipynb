```python
import requests # 서버 요청
from bs4 import BeautifulSoup as bs #html 파싱  
import time # 5초간 휴식 할때 사용  
import pandas as pd # 데이터프레임 
from datetime import datetime # 오늘 연월일 추출 

```

```python
header = {'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"}

jd_list = []

for idxNum in range(1, 71):
    url =f"https://www.saramin.co.kr/zf_user/jobs/list/job-category?page={idxNum}&cat_mcls=2&search_optional_item=n&search_done=y&panel_count=y&preview=y&isAjaxRequest=0&page_count=50&sort=RL&type=job-category&is_param=1&isSearchResultEmpty=1&isSectionHome=0&searchParamCount=1&tab=job-category#searchTitle"
    r = requests.get(url, headers=header)
    soup = bs(r.text, 'lxml')
    job_list = soup.select(".list_item")
    print(f"현재 {idxNum} 페이지 크롤링 중 ",end="\r")

    for i in job_list:
        jd_dict = {}
        link = i.select_one(".job_tit .str_tit")['href']
        title = i.select_one(".col.company_nm .str_tit").text.strip() if i.select_one(".col.company_nm .str_tit").text.strip() is not None else "None"
        sub_title = i.select_one(".col.company_nm .main_corp").text if i.select_one(".col.company_nm .main_corp") is not None else "None" 
        jd = i.select_one(".job_tit span").text if  i.select_one(".job_tit span").text is not None else "None"
        tag = i.select_one(".job_meta").text.strip().split("/") if i.select_one(".job_meta").text.strip().split("/") is not None else  "None"
        badge = i.select_one(".job_badge").text.strip() if i.select_one(".job_badge") is not None else "None" 
        work_place = i.select_one("div.box_item > div.col.recruit_info > ul > li:nth-child(1) > p").text if i.select_one("div.box_item > div.col.recruit_info > ul > li:nth-child(1) > p") is not None else "None"
        career = i.select_one("div.box_item > div.col.recruit_info > ul > li:nth-child(2) > p").text if  i.select_one("div.box_item > div.col.recruit_info > ul > li:nth-child(2) > p") is not None else "None"
        education = i.select_one("div.box_item > div.col.recruit_info > ul > li:nth-child(3) > p").text if  i.select_one("div.box_item > div.col.recruit_info > ul > li:nth-child(3) > p") is not None else "None"

        jd_dict["링크"] = link
        jd_dict["회사명"] = title 
        jd_dict["법인명"] = sub_title
        jd_dict["채용내용"] = jd
        jd_dict["태그"] = tag
        jd_dict["뱃지"] = badge
        jd_dict["근무지"] = work_place
        jd_dict["모집구분"] = career
        jd_dict["학력조건"] = education
        jd_list.append(jd_dict)

    time.sleep(5)  # 5초 대기

    if idxNum == 70:
            break
```
