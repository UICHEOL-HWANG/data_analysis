```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
import time

import pandas as pd 
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install())) #웹 드라이버 버전에 맞게 다운 받고 옵션을 추가해줌 
driver.set_window_size(1920,1080)# 웹 브라우저 해상도 조절 
driver.get("https://www.jobkorea.co.kr/recruit/joblist?menucode=duty&dutyCtgr=10031") # 크롤링할 웹사이트 주소 
```

```python
# 스크랩핑한 페이지를 저장할 리스트
all_pages = []

# 페이지 번호 초기화
page_number = 1
nth_child_number = 1

while page_number <= 50:  # 50 페이지까지 크롤링하도록 설정
    
    # 현재 페이지의 HTML을 가져와서 저장
    html_page = driver.page_source
    print(f"현재 child_number : {nth_child_number}")
    print("현재 페이지:", page_number)
    soup = bs(html_page, "lxml")
    all_pages.append(soup)
    time.sleep(4)
    nth_child_number += 1 
    # 페이지 번호 출력 (테스트용)
    
    # nth-child 선택자를 사용하여 해당 페이지로 이동
    try:
        driver.find_element(By.CSS_SELECTOR, f'#dvGIPaging > div > ul > li:nth-child({nth_child_number}) > a').click()
        driver.find_element(By.CSS_SELECTOR, f'#dvHHPaging > div > ul > li:nth-child({nth_child_number}) > a').click()
        driver.find_element(By.CSS_SELECTOR, f'#dvHRPaging > div > ul > li:nth-child({nth_child_number}) > a').click()
    except Exception as e:
        print("페이지 없음")
        pass
    
    page_number += 1
    time.sleep(5)
    # nth-child_number를 업데이트
    
    if page_number % 10 == 0:  
            if page_number == 10:  
                nth_child_number = 1
                driver.find_element(By.CSS_SELECTOR, "#dvGIPaging > div > p > a").click()
                driver.find_element(By.CSS_SELECTOR, "#dvHHPaging > div > p > a").click()
                driver.find_element(By.CSS_SELECTOR, "#dvHRPaging > div > p > a").click()
                page_number += 1
            else:
                try:
                    nth_child_number = 1
                    driver.find_element(By.CSS_SELECTOR, "#dvGIPaging > div > p:nth-child(3) > a").click()
                    driver.find_element(By.CSS_SELECTOR, "#dvHHPaging > div > p:nth-child(3) > a").click()
                    driver.find_element(By.CSS_SELECTOR, "#dvHRPaging > div > p:nth-child(3) > a").click()
                    page_number += 1
                except Exception as e:
                    print("예외처리",e)
                    pass

            time.sleep(5)  # 페이지 이동 후 잠시 대기 (필요한 경우 조정)
    
# 웹 드라이버 종료
driver.quit()
```

```python
html_page = driver.page_source
soup = bs(html_page, "lxml")
all_pages.append(soup)
```

```python
driver.find_element(By.CSS_SELECTOR,"#dvGIPaging > div > p > a")
```

```python
data_list = [] 

for page_soup in all_pages:
    rcr_cnt_elements = page_soup.select(".rcr_cnt")
    for rcr_cnt in rcr_cnt_elements:
        devloopArea_elements = rcr_cnt.select(".devloopArea")
        for page in devloopArea_elements:
            corp_name = page.select(".tplCo > .link.normalLog")
            titles = page.select(".tplTit > .titBx > strong > a")
            tag = page.select(".tplTit > .titBx > .etc")
            desc = page.select(".tplTit > .titBx > .dsc")
            for title,c_name,tags,dsc in zip(titles,corp_name,tag,desc):
                data_dict = {
                    "타이틀": title.text.strip(),
                    "링크": title["href"],
                    "회사명": c_name.text.strip(),
                    "태그": tags.text.strip().replace(" ", ""),
                    "설명": dsc.text.strip()
                }
                data_list.append(data_dict)
```

```python
data_list
```

```python
df = pd.DataFrame(data_list)

df["태그"] = df["태그"].apply(lambda x: x.split("\n"))
max_tags = max([len(i) for i in df["태그"]])

for i in range(max_tags):
    df[f"태그{i+1}"] = df["태그"].apply(lambda x: x[i] if len(x) > i else "")

df
```

```python
df.to_csv("잡코리아2000개.csv",encoding="utf-8")
```
